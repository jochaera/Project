```python
df['fare_amount'] = df['fare_amount'].fillna(df['fare_amount'].mean())
```

```python
print("남은 결측치 수:", df.isna().sum().sum())
```

```python
df = df.dropna(subset=['fare_amount'])
```

```python
numeric_cols = ['passenger_count', 'trip_distance', 'fare_amount', 'tip_amount', 'tolls_amount']
for col in numeric_cols:
    plt.figure(figsize=(6, 3))
    plt.scatter(df.index, df[col], alpha=0.6)
    plt.title(f"{col} - Scatter Plot")
    plt.xlabel("Index")
    plt.ylabel(col)
    plt.grid(True)
    plt.show()
```

```python
df = df[(df['passenger_count'] > 0) & (df['passenger_count'] <= 6)]
df = df[(df['trip_distance'] > 0) & (df['trip_distance'] < 25)]
df = df[(df['fare_amount'] > 0) & (df['fare_amount'] < 200)]
df = df[df['tip_amount'] < 100]
df = df[df['tolls_amount'] <= 20]
```

```python
df.describe()
```

```python
for col in numeric_cols:
    plt.figure(figsize=(6, 3))
    plt.scatter(df.index, df[col], alpha=0.6, color='green')
    plt.title(f"{col} - After Outlier Removal")
    plt.xlabel("Index")
    plt.ylabel(col)
    plt.grid(True)
    plt.show()
```

```python
# 1. datetime 형식으로 변환
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])

# 2. 주행 시간 (분 단위) 계산
df['trip_duration'] = (df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']).dt.total_seconds() / 60

# 3. 주행 시간이 0 이하인 데이터 제거 (이상치 가능성 있음)
df = df[df['trip_duration'] > 0]
```

```python
# 요금, 거리, 시간 간 상관계수 보기
df[['fare_amount', 'trip_distance', 'trip_duration']].corr()
```

```python
sns.scatterplot(x='trip_duration', y='trip_distance', data=df)
plt.title("Trip Duration vs Distance")
plt.xlabel("Trip Duration (minutes)")
plt.ylabel("Trip Distance (km)")
plt.grid(True)
plt.show()
```

```python
sns.scatterplot(x='trip_duration', y='fare_amount', data=df)
plt.title("Trip Duration vs Fare Amount")
plt.xlabel("Trip Duration (minutes)")
plt.ylabel("Fare Amount ($)")
plt.grid(True)
plt.show()
```
