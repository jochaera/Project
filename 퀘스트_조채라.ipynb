```python
#[메인퀘스트 2번: 간단한 자동사냥 RPG 만들기] 
#간단한 자동사냥 RPG 게임을 만들어 봅시다! 
#사용자의 이름을 입력받아 플레이어를 생성하고, 몬스터들을 차례로 자동 사냥하는 게임을 만들어 보고자 합니다.
```

```python
# Character 클래스 만들기
# 이름, 레벨, 체력, 공격력, 방어력의 속성을 가짐
import random

class Character:
    def __init__(self, name, level, health, attack, defense):
        self.name = name
        self.level = level
        self.health = health
        self.attack = attack
        self.defense = defense
        
#인스턴스의 현재 체력이 0 이상인지
#bool 값을 반환하는 is_alive 메서드 만들기

    def is_alive(self):
        return self.health > 0

#공격을 받았을 때, 
#(받은데미지-본인의방어력)만큼 현재 체력이 감소하는 take_damage 메서드 만들기
#본인의 방어력이 데미지보다 크다면 체력 감소 하지 않음

    def take_damage(self, damage):
        if damage <= self.defense:
            print(f"{self.name}은 공격을 막아냈습니다! (방어력: {self.defense})")
            return   # 함수 종료
        # 공격이 성공했을 때만 실행
        reduced = damage - self.defense
        self.health = max(0, self.health - reduced)
        print(f"{self.name}이 {reduced}의 데미지를 입었습니다! (남은 HP: {self.health})")

#타겟에게 데미지를 입히는 attack_target 메서드 만들기
#데미지는 1부터 공격력 사이의 랜덤한 정수
    
    def attack_target(self, target):
        damage = random.randint(1, self.attack)
        print(f"{self.name}가 {target.name}을(를) 공격합니다! 데미지: {damage}")
        target.take_damage(damage)
```

```python
#Character 클래스를 상속받는 Player와 Monster 클래스를 만들기
#Player 클래스
#Character를 상속받기
#레벨1, 체력100, 공격력25, 방어력5로 초기화 하기

class Player(Character):
    def __init__(self,name):
        super().__init__(name, 1, 100, 25, 5)
        self.experience = 0  # Player 클래스는 경험치 속성을 추가로 가짐
        
        
#인수로 받은 정수만큼 경험치를 획득하는 gain_experience 메서드 만들기

    def gain_experience(self, amount):
        self.experience += amount

    
#현재 경험치가 50이상이면 레벨을1, 공격력을10, 방어력을 5씩 올리는 level_up 메서드 만들기

    def level_up(self):
        if self.experience >= 50:
            self.level += 1
            self.attack += 10
            self.defense += 5
            self.experience -= 50
```

```python
#Monster 클래스
#Character를 상속받기
#몬스터의 레벨에 비례하는 체력, 공격력, 방어력 초기화 하기
#체력: 10~30 사이의 랜덤한 정수* 레벨
#공격력: 5~20 사이의 랜덤한 정수* 레벨
#방어력: 1~5 사이의 랜덤한 정수* 레벨

import random

class Monster(Character):
    def __init__(self, name, level):
        health = random.randint(10, 30) * level
        attack = random.randint(5, 20) * level
        defense = random.randint(1, 5) * level
        super().__init__(name, level, health, attack, defense)
```

```python
#battle 함수 만들기
#Player 인스턴스와 Monster 인스턴스를 인수로 받아 둘 중 하나의 체력이 0 이하가 될 때 까지 공격을 주고받는 함수

def battle(player, monster):
    while player.is_alive() and monster.is_alive():
        player.attack_target(monster)
        if monster.is_alive():
            monster.attack_target(player)

    if player.is_alive():  #만약 Player 인스턴스가 살아남았다면
        player.experience += (monster.level * 20)  #Player 인스턴스에 (몬스터레벨* 20)만큼의 경험치를 추가
        player.level_up()    #player의 레벨업 메서드 호출
        print(f"{player.name}가 {player.level}로 레벨업 했습니다! 레벨: {player.level}")
        print(f"{player.name} 전투 승리!")    #‘전투승리!’를 출력
    else:   #Player 인스턴스가 살아남지 못했을 경우
        print(f"{player.name} 전투 패배..")   # ‘전투패배..’를 출력하기
```

```python
#main 함수만들기

def main():
    monster_dict = {'슬라임': 1, '고블린': 2, '오크': 3}

    name = input("플레이어 이름을 입력하세요: ")
    player = Player(name)

    for monster_name, monster_level in monster_dict.items():
        print("\n" + "="*40)
        print(f"⚔️ {monster_name} 등장! (레벨 {monster_level})")
        print("="*40)
        monster = Monster(monster_name, monster_level)
        battle(player, monster)
        if not player.is_alive():
            print("게임 오버!")
            break
            
main()
```

```python
#Account 클래스

import random

class Account:
    def __init__(self, owner, balance):
        self.bank = "SC은행"
        self.owner = owner
        self.account_number = self.generate_account_number()
        self.balance = balance
        
    def generate_account_number(self):
        part1 = str(random.randint(100, 999))
        part2 = str(random.randint(10, 99))
        part3 = str(random.randint(100000, 999999))
        return f"{part1}-{part2}-{part3}"
```

```python
#클래스 변수
#클래스 변수를 사용해서 Account 클래스로부터 생성된 계좌객체의 개수를 저장하세요.

class Account:
    account_count = 0
    
    def __init__(self, owner, balance):
        Account.account_count += 1   # 계좌가 만들어질 때마다 1 증가
```

```python
#클래스 변수 출력

@classmethod
def get_account_num(cls):
    return cls.account_count
```

```python
#입금 메서드

def deposit(self, amount):
    if amount >= 1:
        self.balance += amount
        return True
    else:
        print("입금은 최소 1원 이상만 가능합니다.")
        return False

#.이자 지급하기
#입금 횟수가 5회가 될 때 잔고를 기준으로 1%의 이자가 잔고에 추가
    if self.deposit_count == 5:
        interest = self.balance * 0.01
        self.balance += interest
        self.deposit_count = 0
```

```python
#출금 메서드

def withdraw(self, amount):
    if amount <= self.balance:
        self.balance -= amount
        return True
    else:
        print("잔액이 부족합니다.")
        return False
```

```python
import random

class Account:
    # Q2. 클래스 변수 - 계좌 개수 추적
    account_count = 0

    def __init__(self, owner, balance):
        self.bank = "SC은행"
        self.owner = owner
        self.balance = balance
        self.account_number = self.generate_account_number()

        # Q10. 입출금 내역 기록
        self.deposit_log = []
        self.withdraw_log = []

        # Q7. 입금 횟수 추적
        self.deposit_count = 0

        Account.account_count += 1  # 계좌 개수 증가

    # Q1. 계좌번호 자동 생성 (3-2-6 자리)
    def generate_account_number(self):
        part1 = str(random.randint(100, 999))
        part2 = str(random.randint(10, 99))
        part3 = str(random.randint(100000, 999999))
        return f"{part1}-{part2}-{part3}"

    # Q6. 계좌 정보 출력
    def display_info(self):
        formatted = format(self.balance, ',')
        print(f"은행이름: {self.bank}, 예금주: {self.owner}, 계좌번호: {self.account_number}, 잔고: {formatted}원")

    # Q4. 입금 메서드
    def deposit(self, amount):
        if amount >= 1:
            self.balance += amount
            self.deposit_log.append(amount)
            self.deposit_count += 1

            # Q7. 이자 지급 (5회마다 1%)
            if self.deposit_count % 5 == 0:
                interest = self.balance * 0.01
                self.balance += interest
                print(f"이자 {int(interest):,}원이 지급되었습니다!")
        else:
            print("입금은 최소 1원 이상만 가능합니다.")

    # Q5. 출금 메서드
    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            self.withdraw_log.append(amount)
        else:
            print("잔액이 부족합니다.")

    # Q3. 클래스 메서드로 전체 계좌 수 조회
    @classmethod
    def get_account_num(cls):
        return cls.account_count

    # Q10. 입금 내역 출력
    def deposit_history(self):
        formatted = [f"{amt:,}원" for amt in self.deposit_log]
        print(f"입금 내역: {formatted}")

    # Q10. 출금 내역 출력
    def withdraw_history(self):
        formatted = [f"{amt:,}원" for amt in self.withdraw_log]
        print(f"출금 내역: {formatted}")
```

```python
# 인스턴스 생성
acc1 = Account("채루", 50000000)
acc2 = Account("아이펠", 1200000)
acc3 = Account("모두연", 800000)
acc4 = Account("M2", 200000)

accounts = [acc1, acc2, acc3, acc4]

# 조건에 맞는 계좌 정보 출력
for account in accounts:
    if account.balance >= 1000000:
        account.display_info()

# 입출금 테스트
acc = Account("채루", 500000)
acc.deposit(110760)
acc.deposit(2005000)
acc.deposit(4800000)
acc.withdraw(150700)
acc.withdraw(10700)

acc.deposit_history()
acc.withdraw_history()
```
