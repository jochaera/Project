```python
#[힌트] cars.csv의 title 변수로부터 브랜드 데이터를 추출할 수 있습니다.
cars['brand_name'] = cars['title'].str.split().str[0].str.lower()
brand['brand_name'] = brand['title'].str.lower()
```

```python
#1.cars.csv와 brand.csv의 데이터 합치기
merged = pd.merge(cars, brand[['brand_name', 'country']], on='brand_name', how='left')
```

```python
merged.head(3)
```

```python
#2.카테고리형 변수를 숫자 형태로 변환하기
#[힌트] Engine, Emission Class 변수에 주목하세요!
merged['Engine'] = merged['Engine'].str.split('L').str[0].astype(float)
merged['Emission Class'] = merged['Emission Class'].str.split().str[1].astype(float)
```

```python
merged.isnull().sum()
```

```python
merged.isnull().mean()
```

```python
merged.describe()
```

```python
#3.결측치 처리하기
#Service history 변수의 결측치는 'Unknown'으로 채우고
merged['Service history'] = merged['Service history'].fillna('Unknown')
```

```python
missing_count = merged.isnull().sum(axis=1)
```

```python
#결측치 변수가 일정 개수 이상 포함된 데이터는 제거합니다.
merged_filtered = merged[missing_count < 5].copy()
```

```python
po_median = merged_filtered['Previous Owners'].median()
engine_median = merged_filtered['Engine'].median()
doors_median = merged_filtered['Doors'].median()
seats_median = merged_filtered['Seats'].median()
emission_median = merged_filtered['Emission Class'].median()
```

```python
merged_filtered['Previous Owners'] = merged_filtered['Previous Owners'].fillna(po_median)
merged_filtered['Engine'] = merged_filtered['Engine'].fillna(engine_median)
merged_filtered['Doors'] = merged_filtered['Doors'].fillna(doors_median)
merged_filtered['Seats'] = merged_filtered['Seats'].fillna(seats_median)
merged_filtered['Emission Class'] = merged_filtered['Emission Class'].fillna(emission_median)
```

```python
#4.Scaler 활용해보기
from sklearn.preprocessing import StandardScaler, RobustScaler, MinMaxScaler
ss = StandardScaler()
rs = RobustScaler()
mm = MinMaxScaler()
```

```python
numeric_cols = ['Price', 'Mileage(miles)', 'Registration_Year',
                'Previous Owners', 'Engine', 'Doors', 'Seats']
```

```python
merged_filtered[numeric_cols] = mm.fit_transform(merged_filtered[numeric_cols])
```

```python
#5.기존 데이터의 정보 70% 이상을 가지는 수준에서 최소한의 주성분 추출 (PCA)
from sklearn.decomposition import PCA

# 1. PCA 대상: 이미 스케일링된 수치형 컬럼들
X = merged_filtered[numeric_cols]

# 2. PCA 객체 만들기 (n_components=0.7 → 전체 정보 중 70% 이상 유지)
pca = PCA(n_components=0.7)

X_pca = pca.fit_transform(X)

num_components = X_pca.shape[1]  # 몇 개의 축(차원)?
explained_variance_ratio = pca.explained_variance_ratio_.sum()
```

```python
print(f"추출된 주성분 수: {num_components}")
print(f"설명된 전체 분산 비율: {explained_ratio:.4f}")
```

```python
#국가별 총 브랜드 개수는?
# 1. 국가별로 그룹화해서 브랜드의 고유 개수 세기
brand_counts_by_country = merged_filtered.groupby("country")["brand_name"].nunique().reset_index()

# 2. 컬럼 이름 정리 (country, 브랜드 개수)
brand_counts_by_country.columns = ["country", "num_unique_brands"]

# 3. 브랜드 개수 기준으로 내림차순 정렬
brand_counts_by_country = brand_counts_by_country.sort_values(by="num_unique_brands", ascending=False)

# 4. 결과 출력
print(brand_counts_by_country)
```

```python
#상관관계가 높은 변수들이 있는지 확인해보세요.
correlation_matrix = merged_filtered[numeric_cols].corr()
```
